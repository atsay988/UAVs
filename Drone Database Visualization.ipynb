```python
from tqdm import tqdm
from urllib.parse import urljoin, urlparse
import matplotlib.pyplot as plt
import re
import numpy as np
```

```python
import requests
from bs4 import BeautifulSoup

headers = {
    'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10.12; rv:55.0) Gecko/20100101 Firefox/55.0',
}

URL = 'http://drones.cnas.org/drones/'
page = requests.get(URL, headers = headers)
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
drone_list = soup.find_all('div',class_='drone-box')
drone_list


#download images
img_list = soup.find_all('div', class_="drone-name has-thumbnail")
img_list

img_urls = []

for img in img_list:
    url = img.get('style')
    img_urls.append(url)

img_urls_new = []

for item in img_urls:
    item = item[22:-3]
    img_urls_new.append(item)
```

```python
name_list = soup.find_all('h2')
```

```python
## identify drone
### find h2 and all dd elements

drone_dict = {}

for drone in drone_list:
    keys = (drone.find('h2')).text
    data = drone.find_all('dd')
    drone_dict[keys] = (data[0].text, data[1].text, data[3].text, data[4].text, data[5].text, data[6].text, data[7].text, data[8].text, data[9].text, data[10].text)


```

```python
#next steps
#create a "drone" class w/ name, country, company, endurance, range, payload cap, max speed, MTOW, width, length

#parse each drone in drone list
#h2 = name
#

fields = ("platform", "country", "company", "endurance", "range", "payload_cap", "max_speed", "ceiling", "MTOW", "width", "length", "image_url")

class uav:
    def __init__(self, country, company, platform, endurance, range_, payload_cap, max_speed, ceiling, MTOW, width, length):
        self.country = country
        self.company = company
        self.platform = platform
        self.endurance = endurance
        self.range_ = range_
        self.payload_cap = payload_cap
        self.max_speed = max_speed
        self.ceiling = ceiling
        self.MTOW = MTOW
        self.width = width
        self.length = length


uavs = []
        
for drone in drone_list:
    data = drone.find_all('dd')
    uavs.append(uav(data[0].text, data[1].text, data[2].text, re.sub("[^0-9]","", data[3].text), re.sub("[^0-9]","", data[4].text), re.sub("[^0-9]","", data[5].text), re.sub("[^0-9]","", data[6].text), re.sub("[^0-9]","",data[7].text), re.sub("[^0-9]","", data[8].text), re.sub("[^0-9]","", data[9].text), re.sub("[^0-9]","", data[10].text)))

    
uavs.sort(key=lambda x: x.endurance)
#sorted(uavs, key=lambda x: x.endurance)

plt.figure(figsize=(200,70), facecolor='grey') # figure settings

for x in range(len(uavs)):
    
    if len(uavs[x].endurance) == 0:
        uavs[x].endurance = 0
    uavs[x].endurance = int(uavs[x].endurance)
    
    if len(uavs[x].range_) == 0:
        uavs[x].range_ = 0
    uavs[x].range_ = int(uavs[x].range_)
        
    if len(uavs[x].payload_cap) == 0:
        uavs[x].payload_cap = 0
    uavs[x].payload_cap = int(uavs[x].payload_cap)
    
    if len(uavs[x].max_speed) == 0:
        uavs[x].max_speed = 0
    uavs[x].max_speed = int(uavs[x].max_speed)
    
    if len(uavs[x].ceiling) == 0:
        uavs[x].ceiling = 0
    uavs[x].ceiling = int(uavs[x].ceiling)
        
    if len(uavs[x].MTOW) == 0:
        uavs[x].MTOW = 0
    uavs[x].MTOW = int(uavs[x].MTOW)
        
    if len(uavs[x].width) == 0:
        uavs[x].width = 0
        
    if len(uavs[x].length) == 0:
        uavs[x].length = 0


# Create scatter plot
    plt.bar(uavs[x].platform, uavs[x].ceiling)
    plt.xticks(rotation = 90, fontsize = 65)
    plt.yticks(fontsize = 80)
    plt.xlabel('UAV', fontsize=100)
    plt.ylabel('Ceiling (m)', fontsize=100)
    plt.ylim(0,25000)




```
